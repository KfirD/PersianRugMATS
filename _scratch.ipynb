```python
from utils import model_scan
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_sparse = 500
ratio_min = 5/500
ratio_max = 500*0.6
num_points = 30
ratios = np.linspace(ratio_min, ratio_max, num_points)
p_feats = [0.03]

experiment_name = "nonlinear_special"
try:
    nonlinear_models = model_scan.load_trained_models(experiment_name)
except:
    nonlinear_models = model_scan.train_multiple_models(ratios, p_feats, n_sparse, non_linearity = True, experiment_name=experiment_name)
final_losses_nonlinear = [model.final_loss() for model in nonlinear_models]
```

```python
plt.plot(ratios, final_losses_nonlinear)
```
